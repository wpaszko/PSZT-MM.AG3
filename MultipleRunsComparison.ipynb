```python
import creators
import creatures
import crossovers
import evaluators
import evolution
import mutators
import selects
import logger
```

```python
a = sum([1, 3, 5, 10])
b = evaluators.get_product_of_list([2, 4, 6, 7, 8, 9])
n = 10
```

```python
creature_class = creatures.DeckCreature
creator = creators.RandomCreator(creature_class, n)
evaluator = evaluators.LogarithmicMeanDeckEvaluator(a, b)
selector = selects.ExponentialSelection(1.2)
crossover = crossovers.RandomIndependentCrossover(creature_class)
mutator = mutators.RandomIndependentSwapMutator(creature_class, 0.05)

population_size = 10
```

```python
results = []

for i in range(10):
    evol = evolution.Evolution(creator,
                           evaluator,
                           selector,
                           crossover,
                           mutator,
                           population_size)
    
    log = logger.Logger()
    best = evol.evolve_until_satisfied(1.0, log)
    results.append((best, log))
```

```python
print("Set values:")
print("A: ", a)
print("B: ", b)
print()

for best, log in results:
    print("Generations: ", log.current_gen)
    print("End fitness: ", log.lowests[-1])
    print("End matching: ", log.matchings[-1])
    print()
    print("A: ", best.get_stack(True), ", sum: ", evaluators.get_sum_on_stack(best, True))
    print("B: ", best.get_stack(False), ", product: ", evaluators.get_product_on_stack(best, False))
    log.plot_mean(figsize=(4,4))
```
